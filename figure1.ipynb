```python
from matplotlib import pyplot as plt
from matplotlib.patches import FancyBboxPatch

# 그림 생성 함수
def draw_flowchart():
    fig, ax = plt.subplots(figsize=(8, 6))  # 캔버스 크기 설정

    # 데이터 처리 단계
    ax.add_patch(FancyBboxPatch((0.3, 0.7), 0.4, 0.1, boxstyle="round,pad=0.1", edgecolor="black", facecolor="lightblue"))
    ax.text(0.5, 0.75, "Data Preprocessing", fontsize=12, ha="center", va="center")

    # 모델 학습 단계
    ax.add_patch(FancyBboxPatch((0.3, 0.5), 0.4, 0.1, boxstyle="round,pad=0.1", edgecolor="black", facecolor="lightgreen"))
    ax.text(0.5, 0.55, "BERT Fine-tuning", fontsize=12, ha="center", va="center")

    # 감정 예측 단계
    ax.add_patch(FancyBboxPatch((0.3, 0.3), 0.4, 0.1, boxstyle="round,pad=0.1", edgecolor="black", facecolor="lightpink"))
    ax.text(0.5, 0.35, "Sentiment Prediction", fontsize=12, ha="center", va="center")

    # 화살표
    ax.annotate("", xy=(0.5, 0.7), xytext=(0.5, 0.6), arrowprops=dict(arrowstyle="->"))
    ax.annotate("", xy=(0.5, 0.5), xytext=(0.5, 0.4), arrowprops=dict(arrowstyle="->"))

    # 축 설정
    ax.axis("off")  # 축 제거
    plt.title("Sentiment Analysis Flow", fontsize=16, ha="center")  # 제목 중앙 정렬
    plt.show()

draw_flowchart()

```
