```python
import matplotlib.pyplot as plt

# 예시 데이터: 오분류된 텍스트 수
misclassification_counts = {
    "Neutral → Negative": 120,
    "Negative → Neutral": 80,
    "Positive → Negative": 50,
    "Positive → Neutral": 30,
    "Neutral → Positive": 20,
    "Negative → Positive": 15,
}

# 클래스와 값 추출
categories = list(misclassification_counts.keys())
values = list(misclassification_counts.values())

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
plt.barh(categories, values, color="skyblue")
plt.xlabel("Number of Misclassified Texts")
plt.ylabel("Misclassification Type")
plt.title("Figure 3: Number of Misclassified Texts by Emotion Class")
plt.tight_layout()
plt.show()
```
